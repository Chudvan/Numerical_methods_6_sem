```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
e = math.e
t = np.linspace(0, 1, 101)               
x = -4 * e ** t + 3* e **(2 * t) + np.cos(t) - 2 * np.sin(t)     
y = -4 * e ** t + 2* e **(2 * t) + 2*np.cos(t) - 2 * np.sin(t)    

fig, ax = plt.subplots()                      
ax.plot(t, y, color="blue", label="y(t)")     
ax.plot(t, x, color="red", label="x(t)")  
ax.set_xlabel("t")                       
ax.set_ylabel("x / y")                    
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='green', label='y(x)')
ax.set_xlabel("x")  
ax.set_ylabel("y")                              # подпись у вертикальной оси y
ax.legend()
```

```python
e = math.e
t = np.linspace(0, 3, 31)               
x = (1/2) * e ** (8 * t) - 1/2 - 6 * t + 1    
y = (3/4) * e ** (8 * t) + 1/4 + 3 * t    

fig, ax = plt.subplots()                      
ax.plot(t, y, color="blue", label="y(t)")     
ax.plot(t, x, color="red", label="x(t)")  
ax.set_xlabel("t")                       
ax.set_ylabel("x / y")                    
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='green', label='y(x)')
ax.set_xlabel("x")  
ax.set_ylabel("y")                              # подпись у вертикальной оси y
ax.legend()
```
