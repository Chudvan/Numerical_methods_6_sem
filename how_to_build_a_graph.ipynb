```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 0.15, 4)
```

```python
x
```

```python
y1 = np.array(x + 1)
```

```python
y2 = np.array([1, 1.05, 1.0931, 1.1347])
```

```python
y3 = np.array([1, 1.0477, 1.0912, 1.1311])
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label='Taylor')
ax.plot(x, y2, color="red", label="Euler")
ax.plot(x, y3, color="green", label="Runge-Kutta")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```
